
# Comparative Analysis of Numbers from 1-100 Across Languages

This project explores how numbers from 0 to 100 are named in different languages by analyzing the starting letters of each number name. The goal is to visually compare these starting letters across languages to identify similarities and differences in linguistic patterns. By examining these patterns, we hope to gain insights into the linguistic structure and evolution of numeral systems across different language families.


```python

import pandas as pd
import matplotlib.pyplot as plt
import string
```

```python

# Load the dataset
csv_path = '0-100.csv'
data = pd.read_csv(csv_path, delimiter=';')
# Clean the data to focus on the starting letters of each word
data_cleaned = data.applymap(lambda x: x[0].lower() if x[0].lower() in string.ascii_lowercase else 'other')
```

```python

# Define colors for each language
language_colors = {
    'English': '#1f77b4', 'French': '#ff7f0e', 'Italian': '#2ca02c', 
    'German': '#d62728', 'Portuguese': '#9467bd', 'Spanish': '#8c564b',
    'Russian': '#e377c2', 'Japanese': '#7f7f7f', 'Mandarin': '#bcbd22'
}

# Plot to compare starting letters across all languages
plt.figure(figsize=(20, 10))
for language in data.columns:
    letter_counts = data_cleaned[language].value_counts().reindex(list(string.ascii_lowercase), fill_value=0)
    plt.plot(list(string.ascii_lowercase), letter_counts, label=language, color=language_colors[language], marker='o')
plt.title('Comparison of Starting Letters in Different Languages')
plt.xlabel('Starting Letters')
plt.ylabel('Frequency of Each Letter')
plt.legend(title="Languages")
plt.grid(True)
plt.show()
```

```python

# Function to create pairwise comparison plots for two languages
def create_pairwise_plot(language1, language2):
    fig, ax = plt.subplots(figsize=(15, 10))
    starting_letters = list(string.ascii_lowercase)
    
    # Plot for the first language
    for i, entry in enumerate(data_cleaned[language1]):
        y_position = starting_letters.index(entry) if entry in starting_letters else -1
        ax.scatter(i - 0.2, y_position, color=language_colors[language1], marker='o', label=language1 if i == 0 else '')

    # Plot for the second language
    for i, entry in enumerate(data_cleaned[language2]):
        y_position = starting_letters.index(entry) if entry in starting_letters else -1
        ax.scatter(i + 0.2, y_position, color=language_colors[language2], marker='o', label=language2 if i == 0 else '')

    ax.set_xlabel('Number (0-100)')
    ax.set_yticks(range(26))
    ax.set_yticklabels(starting_letters)
    ax.set_title(f'Comparison of Starting Letters between {language1} and {language2}')
    ax.legend()
    plt.show()

# Generate plots for multiple language pairs
language_pairs = [
    ('English', 'French'), ('German', 'Italian'), ('Russian', 'Portuguese'), 
    ('Spanish', 'Mandarin'), ('French', 'Italian'), ('Portuguese', 'Spanish'),
    ('English', 'German'), ('Russian', 'English'), ('Japanese', 'Mandarin'),
    ('Russian', 'German'), ('Italian', 'Spanish'), ('French', 'Portuguese'),
    ('English', 'Japanese'), ('German', 'Mandarin'), ('Spanish', 'Portuguese'),
    ('English', 'Mandarin')
]
for lang1, lang2 in language_pairs:
    create_pairwise_plot(lang1, lang2)
```

**Understanding Linguistic Patterns Through Numerals**

This project embarked on a fascinating exploration of how different languages express the simplest of concepts—numbers. ### Commonalities Across Language Families

Our analysis revealed significant similarities within certain language families. Such findings affirm the hypothesis that linguistic evolution within language families preserves certain phonetic attributes.

### Distinctive Linguistic Features

Conversely, the study highlighted stark differences when comparing languages from unrelated families. For example, Japanese and English exhibited almost entirely distinct patterns, which underscores the profound impact of linguistic structure and historical development on language. Japanese, with its unique phonetic inventory and syllabic structure, starts number names with sounds that are rarely used in English.

### Unexpected Anomalies

Some interesting anomalies were observed, such as the frequent use of vowels as starting letters in some languages but not others. These anomalies provide an intriguing avenue for further research, possibly exploring how phonological rules in languages influence the formulation of number names.

The findings from this project contribute to our understanding of phonological organization and its implications for linguistic theory.
